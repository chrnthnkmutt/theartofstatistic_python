#### Synopsis

In the introduction of the book, David Spiegelhalter discusses the chilling case of Harold Shipman, a British doctor who became one of the most prolific serial killers in history. Shipman was responsible for the deaths of many of his patients through the administration of lethal doses of drugs. This chapter serves as an opening illustration of the importance of statistics in uncovering patterns and anomalies in data.

Spiegelhalter uses the Shipman case to introduce the concept of "prior probability" and how it played a significant role in the investigation. Prior probability is the likelihood of an event occurring before new evidence is taken into account. In the case of Shipman, it was highly improbable that so many patients under his care would die of natural causes, and this raised suspicions that eventually led to his arrest and conviction.

##### Age and Year of Shipman Victims, including gender

The data below are Harold Shipman's patients that he fails on overdosed injection, which contains 215 patients in the list, containing in [00-1-shipman-confirmed-victims-x.csv](./00-1-shipman-confirmed-victims-x.csv), and taken from [Chronological List of Decided Cases in the Shipman Inquiry](https://webarchive.nationalarchives.gov.uk/20090808221518/http://www.the-shipman-inquiry.org.uk/fr_casesbyyear.asp?year=74&from=r).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("00-1-shipman-confirmed-victims-x.csv")

s = sns.scatterplot(data=df, x="fractionalDeathYear", y="Age", hue="gender")
s.set(xlabel="Year", ylabel="Age of victim")
s.set(xlim=(1974, 1998), ylim=(39, 95))
s.set_xticks(range(1975, 1996, 5))
s.set_yticks(range(40, 100, 10))

sns.set_palette("Set1")
sns.despine()
df['gender'] = df['gender'].replace({'F': 0, 'M': 1})

g = sns.JointGrid(data=df, x="fractionalDeathYear", y="Age")
g = g.plot(sns.histplot, sns.histplot)

# sns.histplot(data=df, x="fractionalDeathYear", y="Age", ax=s)

plt.show()
```

Figure above showing the age and the year of death of Harold Shipman's 215 confirmed victims. Bar-charts have been added on the axes to reveal the patterns of ages and the pattern of years in which he committed murders. The bar-charts are called marginal distributions, and the plot is called a scatterplot.

##### At what time of day did Harold Shipmans victims die?
Data are contained in [0-2-shipman-times-x.csv](./00-2-shipman-times-x.csv), and taken from [Harold Shipmans Clinical Practice 1974-1998: A Clinical Audit Commissioned by the Chief Medical Officer, by Richard Baker, page 40, Figure 5.2](https://webarchive.nationalarchives.gov.uk/20090808160000/http://www.the-shipman-inquiry.org.uk/ge_doclist.asp?ID=5), derived from the cremation forms of Shipmans victims.

```python
import pandas as pd
import matplotlib.pyplot as plt

shipmantimes = pd.read_csv("00-2-shipman-times-x.csv")
Hour = shipmantimes['Hour']
Shipman = shipmantimes['Shipman']
Comparison = shipmantimes['Comparison']

plt.plot(Hour+0.5, Shipman, 'r-', linewidth=2, label="Shipman")
plt.plot(Hour+0.5, Comparison, 'b--', linewidth=2, label="comparison GPs")
plt.xlabel("hour")
plt.ylabel("Percentage of deaths")
plt.ylim(0, 16)
plt.legend()
plt.show()
```

Figure 0.2 - The time at which Harold Shipman's patients died, compared to the times at which patients of other local general practitioners died. The pattern does not require sophisticated statistical analysis.

#### Summary

The case of Harold Shipman is used as a real-life example to illustrate the importance of statistics and data analysis. Harold Shipman was a British doctor who was convicted of being a serial killer. Here is a summary of how his case is discussed in the book:

1. Serial Killer Doctor: David Spiegelhalter presents Harold Shipman as a disturbing case of a doctor who betrayed the trust of his patients. Shipman was a family practitioner who was convicted of killing at least 218 of his patients, although the actual number of victims may have been even higher.

2. Uncovering the Pattern: The book discusses how the authorities began to investigate Shipman after noticing an unusual pattern of deaths among his patients. Typically, deaths in a medical practice would be spread out over time, but in Shipman's case, there was an alarming concentration of deaths.

3. Statistical Analysis: Spiegelhalter introduces the concept of prior probability, which is the likelihood of an event happening before considering new evidence. In Shipman's case, the prior probability that so many patients would die of natural causes in such a short time frame was incredibly low, raising suspicion.

4. Role of Statistics: The Shipman case demonstrates how statistical analysis played a crucial role in uncovering the anomaly and ultimately led to his arrest and conviction. Statistics and data analysis were essential tools for identifying the unusual pattern of patient deaths and for building a case against Shipman.

5. Lessons in Data Analysis: The case serves as a powerful example of how statistics can be used to detect anomalies and patterns in data, even in situations that are initially difficult to detect. It highlights the importance of data analysis and critical thinking in various fields, including medicine and law enforcement.

Overall, the Harold Shipman case is used in "The Art of Statistics" to emphasize the role of statistics in uncovering unexpected patterns and anomalies in data, and how this analytical approach can have real-world implications, including in the most troubling and unsettling cases.
