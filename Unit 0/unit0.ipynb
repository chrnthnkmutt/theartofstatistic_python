The data below are Harold Shipman's patients that he fails on overdosed injection, which contains 215 patients in the list, containing in 00-1-shipman-confirmed-victims-x.csv, and taken from Chronological List of Decided Cases in the Shipman Inquiry.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("00-1-shipman-confirmed-victims-x.csv")

s = sns.scatterplot(data=df, x="fractionalDeathYear", y="Age", hue="gender2")
s.set(xlabel="Year", ylabel="Age of victim")
s.set(xlim=(1974, 1998), ylim=(39, 95))
s.set_xticks(range(1975, 1996, 5))
s.set_yticks(range(40, 100, 10))

sns.set_palette("Set1")
sns.despine()

g = sns.JointGrid(data=df, x="fractionalDeathYear", y="Age")
g = g.plot(sns.histplot, sns.histplot)

# sns.histplot(data=df, x="fractionalDeathYear", y="Age", ax=s)

plt.show()
```

Figure above showing the age and the year of death of Harold Shipman's 215 confirmed victims. Bar-charts have been added on the axes to reveal the patterns of ages and the pattern of years in which he committed murders. The bar-charts are called marginal distributions, and the plot is called a scatterplot.

```python
import pandas as pd
import matplotlib.pyplot as plt

shipmantimes = pd.read_csv("00-2-shipman-times-x.csv")
Hour = shipmantimes['Hour']
Shipman = shipmantimes['Shipman']
Comparison = shipmantimes['Comparison']

plt.plot(Hour+0.5, Shipman, 'r-', linewidth=2, label="Shipman")
plt.plot(Hour+0.5, Comparison, 'b--', linewidth=2, label="comparison GPs")
plt.xlabel("hour")
plt.ylabel("Percentage of deaths")
plt.ylim(0, 16)
plt.legend()
plt.show()
```

Figure 0.2 - The time at which Harold Shipman's patients died, compared to the times at which patients of other local general practitioners died. The pattern does not require sophisticated statistical analysis to be seen.
